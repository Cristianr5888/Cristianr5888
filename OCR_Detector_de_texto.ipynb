```python
import cv2
import easyocr
import matplotlib.pyplot as plt
import numpy as np

#find file
img='/content/osmar_credencial.jpeg'
rimg = cv2.imread(img)
#create the reader
reader = easyocr.Reader(['en'],gpu = False, verbose= False) #lenguaje    example: en -english  or es -español

#detect text
text= reader.readtext(rimg)
thereshold = 0.15

#box detection text
for tx,t in enumerate(text):
  print(t)

  bbox, textP, score = t

  if score > thereshold:        #corners        color
    cv2.rectangle(rimg, bbox[0], bbox[2] , (0,255,0), 2)
    cv2.putText(rimg, textP, bbox[3], cv2.FONT_HERSHEY_COMPLEX, 0.7, (255,0,0),2)

plt.imshow(cv2.cvtColor(rimg, cv2.COLOR_BGR2RGB))
plt.show()
```
